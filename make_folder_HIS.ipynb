```python
import requests
import os
import re

os.makedirs("plays", exist_ok=True)

def sanitize_filename(s):
    s = s.replace("/", "-").replace("\\", "-")
    s = re.sub(r'[:*?"<>|]', '', s)
    s = s.strip().replace(" ", "_")
    return s

for play in urls:
    parts = play.split()
    url = parts[-1]
    year = parts[-2]
    title = " ".join(parts[:-2]).strip()

    safe_name = sanitize_filename(f"{title}_{year}")

    try:
        r = requests.get(url)
        if r.status_code == 200:
            filename = f"plays/{safe_name}.xml"
            with open(filename, "w", encoding="utf-8") as f:
                f.write(r.text)
            print("OK:", safe_name)
        else:
            print("Statusfeil:", safe_name, r.status_code)

    except Exception as e:
        print("Problem med", safe_name, "->", e)
```
